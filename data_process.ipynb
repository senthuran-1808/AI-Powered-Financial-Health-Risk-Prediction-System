```python
master_df.columns = (
    master_df.columns
    .str.strip()
    .str.replace(" ", "_")
    .str.replace("+", "", regex=False)
    .str.replace("%", "Percentage", regex=False)
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
    .str.replace("/", "_", regex=False)
)
```

```python
master_df.columns = master_df.columns.str.strip()
print(master_df.columns)
```

```python
# Create Debt Ratio Formula

# Debt Ratio = Borrowings / Total Assets

# Use to check companies heavily dependent on loans are riskier.

# Higher financial risk

master_df["Debt_Ratio"] = (
    master_df["Borrowings"] /
    master_df["Total_Assets"]
)


```

```python
# Create Reserve Ratio Formula

# Reserve Ratio = Reserves / Total Assets

# Financially stronger company

master_df["Reserve_Ratio"] = (
    master_df["Reserves"] /
    master_df["Total_Assets"]
)
```

```python
# Create Profit Margin Formula

# Profit Margin = Net Profit / Sales * 100

# How efficiently company generates profit

master_df["Profit_Margin"] = (
    master_df["Net_Profit"] /
    master_df["Sales"]
) * 100
```

```python
# Create Asset Turnover Ratio Formula

# Asset Turnover = Sales / Total Asserts 

# Asset utilization efficiency

master_df["Asset_Turnover"] = (
    master_df["Sales"] /
    master_df["Total_Assets"]
)
```

```python
# Create Cash Flow Ratio Formula

# Cashflow Ratio = Cash from Operating Activity / Total Liabilities

# Ability to pay liabilities using operational cash

master_df["Cashflow_Ratio"] = (
    master_df["Cash_from_Operating_Activity"] /
    master_df["Total_Liabilities"]
)
```

```python
# Handling Infinite values

master_df.replace(
    [np.inf, -np.inf],
    np.nan,
    inplace=True
)

numeric_cols = master_df.select_dtypes(
    include=np.number
).columns

master_df[numeric_cols] = master_df[
    numeric_cols
].fillna(0)
```

```python
# Creating Financial Health Score

master_df["Financial_Health_Score"] = (
    (
        master_df["Reserve_Ratio"] * 0.3
    ) +
    (
        master_df["Profit_Margin"] * 0.3
    ) +
    (
        master_df["Cashflow_Ratio"] * 0.2
    ) -
    (
        master_df["Debt_Ratio"] * 0.2
    )
)
```

```python
# Classifing company risk level with high, medium and low

master_df["Risk_Level"] = np.where(
    master_df["Debt_Ratio"] > 0.6,
    "High Risk",
    np.where(
        master_df["Debt_Ratio"] > 0.3,
        "Medium Risk",
        "Low Risk"
    )
)
```

```python
# Create Health Category Score:

def health_category(score):

    if score > 25:
        return "Excellent"

    elif score > 15:
        return "Good"

    elif score > 5:
        return "Moderate"

    else:
        return "Weak"
```

```python
master_df["Health_Category"] = master_df[
    "Financial_Health_Score"
].apply(health_category)
```

```python
# Recommendation Engine based on the finance analysis

def generate_recommendation(row):

    # High Debt Risk
    if row["Debt_Ratio"] > 0.7:
        return "High debt dependency detected. Reduce borrowing exposure and improve cash reserves."

    # Weak Profitability
    elif row["Profit_Margin"] < 5:
        return "Low profitability observed. Improve operational efficiency and revenue growth."

    # Weak Cash Flow
    elif row["Cashflow_Ratio"] < 0.2:
        return "Weak operational cash flow detected. Improve liquidity management."

    # Strong Financial Health
    elif row["Financial_Health_Score"] > 20:
        return "Strong financial condition with stable profitability and reserves."

    # Moderate Condition
    else:
        return "Moderate financial stability observed with balanced financial performance."
```

```python
master_df["AI_Recommendation"] = master_df.apply(
    generate_recommendation,
    axis=1
)
```

```python
final_dashboard_df = master_df[
    [
        "CompanyName",
        "Industry",
        "quarter10yrs",

        "Sales",
        "Net_Profit",
        "Borrowings",
        "Total_Assets",

        "Debt_Ratio",
        "Profit_Margin",
        "Financial_Health_Score",
        "Health_Category",

        "Risk_Level",
        "AI_Recommendation",

        "PE",
        "PB",
        "RONW"
    ]
]
```

```python
## Model training

features = [
    "Sales",
    "Net_Profit",
    "Borrowings",
    "Total_Assets",
    "Debt_Ratio",
    "Profit_Margin",
    "Financial_Health_Score",
    "PE",
    "PB",
    "RONW"
]

target = "Risk_Level"
```

```python
X = final_dashboard_df[features]

y = final_dashboard_df[target]
```

```python
# Converting Categorical values into numerical

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

y_encoded = encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42
)
```

```python
## Linear Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr_model = LogisticRegression(
    max_iter=1000
)

lr_model.fit(X_train, y_train)

lr_pred = lr_model.predict(X_test)

lr_accuracy = accuracy_score(
    y_test,
    lr_pred
)

print("Logistic Regression Accuracy:", lr_accuracy)
```

```python
#Random Forest

from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

rf_model.fit(X_train, y_train)

rf_pred = rf_model.predict(X_test)

rf_accuracy = accuracy_score(
    y_test,
    rf_pred
)

print("Random Forest Accuracy:", rf_accuracy)
```

```python
# Gradient Boosting

from sklearn.ensemble import GradientBoostingClassifier

gb_model = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.05,
    random_state=42
)

gb_model.fit(X_train, y_train)

gb_pred = gb_model.predict(X_test)

gb_accuracy = accuracy_score(
    y_test,
    gb_pred
)

print("Gradient Boosting Accuracy:", gb_accuracy)
```

```python
import pandas as pd

model_results = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Random Forest",
        "Gradient Boosting"
    ],
    "Accuracy": [
        lr_accuracy,
        rf_accuracy,
        gb_accuracy
    ]
})

print(model_results)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.barplot(
    data=model_results,
    x="Model",
    y="Accuracy"
)

plt.title("Model Accuracy Comparison")

plt.show()
```
